```python
import pandas as pd
import matplotlib.pyplot as plt

# ----- グラフ描画のバックエンド設定（VSCodeで表示されない問題を防ぐ） -----
plt.rcParams['figure.figsize'] = (10, 6)

# ----- 棒グラフを作成する関数 -----
def bar_chart(labels, values):
    plt.figure()
    plt.bar(labels, values)
    plt.xticks(rotation=45)
    plt.ylabel("Total Deaths")
    plt.title("Deaths by Disaster Type")
    plt.tight_layout()
    plt.show()

# ----- CSV 読み込み（★★★ここを書き換えてください★★★） -----
df = pd.read_csv(r"C:\Users\liyua\Desktop\your_file.csv")
# ↑ あなたの CSV のフルパスをここに貼ればOK（Explorer → 右クリック → Copy Path）

# ----- 使用する列名 -----
disaster_col = "Disaster Type"
death_col = "Total Deaths"

# ----- 災害タイプごとの死亡者数グルーピング -----
summary = df.groupby(disaster_col)[death_col].sum()

# ----- グラフ用の値作成 -----
labels = summary.index.tolist()
values = summary.values.tolist()

# ----- グラフ描画 -----
bar_chart(labels, values)
```
